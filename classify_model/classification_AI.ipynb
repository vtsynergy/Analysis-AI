```python
import os
import glob
import json
from shutil import copy2,move
```

```python
seg_root = '/claraDevDay/seg_model'
MMAR_ROOT="/claraDevDay/classify_model"
print ("setting MMAR_ROOT=",MMAR_ROOT)
```

```python
# delete old images
olds = glob.glob(MMAR_ROOT+'/dataset_root/valid/*')
for i in olds:
    os.remove(i)
#     !rm $i
# copy nii mask images in dataset
news = glob.glob(seg_root+'/eval/*/*')
dataset_path = MMAR_ROOT+'/dataset_root/valid/'
for i in news:
    copy2(i, dataset_path)
#     !cp $i $dataset_path 
# copy new nii images in dataset
news = glob.glob(seg_root+'/dataset_root/valid/*')
dataset_path = MMAR_ROOT+'/dataset_root/valid/'
for i in news:
    copy2(i, dataset_path)
#     !cp $i $dataset_path 
```

```python
def maskSort(e):
    lists = e.split('/')
    file = lists[len(lists)-1]
    lists_name = file.split('_')
    ID = int(lists_name[1])
#     print(lists_name[0])
    return lists_name[0],ID
```

```python
def niiSort(e):
    lists = e.split('/')
    file = lists[len(lists)-1]
    lists = file.split('.')[0]
    lists = lists.split('_')
    ID = int(lists[len(lists)-1])
#     print(lists[0])
    return lists[0],ID
```

```python
# generate json file
mask_list = glob.glob(MMAR_ROOT+'/dataset_root/valid/*model*')
mask_list.sort(key = maskSort)
nii_list = list(set(glob.glob(MMAR_ROOT+'/dataset_root/valid/*')) - set(glob.glob(MMAR_ROOT+'/dataset_root/valid/*model*')))
nii_list.sort(key = niiSort)

data = {}
data['label_format'] = [2]
data['validation'] = []
# skip_list = [*range(0,31),*range(51,78),*range(149,156),169,170,171,187]
skip_list = []

for i in range(len(mask_list)):
    if i not in skip_list:
        mask_name_list = mask_list[i].split('/')
        nii_name_list = nii_list[i].split('/')
        mask_name = mask_name_list[len(mask_name_list)-1]
        nii_name = nii_name_list[len(nii_name_list)-1]
        data['validation'].append({
            'image': 'valid/'+ nii_name,
            'label_image': '/claraDevDay/classify_model/dataset_root/valid/'+ mask_name
        })
clas_json_path = MMAR_ROOT + '/dataset_root/'
with open(clas_json_path + 'datalist.json', 'w') as outfile:
    json.dump(data, outfile)
```
